```python
# Ensemble Methods in Machine Learning

## Introduction

Ensemble methods are machine learning techniques that combine multiple models (called base learners) to improve the overall prediction performance. Instead of relying on a single model, ensemble methods aggregate the predictions of several models to produce more accurate and stable results. These methods are widely used in applications such as medical diagnosis, fraud detection, recommendation systems, and image recognition.

## Working of Ensemble Methods

The basic working process of ensemble methods includes the following steps:

1. Multiple base models are trained using the same dataset or different subsets of the dataset.
2. Each model makes its own prediction independently or sequentially.
3. The predictions from all models are combined using techniques such as voting or averaging.
4. A final prediction is produced, which is usually more accurate than individual model predictions.

## Types of Ensemble Methods

### 1. Bagging (Bootstrap Aggregating)

Bagging trains multiple models independently using different random subsets of the training dataset. The final output is obtained by averaging predictions (for regression) or voting (for classification).

Example:
Random Forest Algorithm

### 2. Boosting

Boosting trains models sequentially. Each new model focuses on correcting the errors made by the previous model. It improves model performance step by step.

Examples:
AdaBoost
Gradient Boosting
XGBoost

### 3. Stacking

Stacking combines predictions from multiple different machine learning models and uses another model called a
meta-learner to produce the final prediction.

Example:
Combining Decision Tree, Support Vector Machine, and Logistic Regression using a meta-classifier.

## Advantages of Ensemble Methods

* Improves prediction accuracy
* Reduces overfitting problems
* Provides better stability and robustness
* Works well with complex datasets
* Useful in healthcare applications such as heart disease prediction

## Disadvantages of Ensemble Methods

* Requires more computational power
* Training time is higher compared to single models
* Difficult to interpret results
* Requires more memory
* Implementation is more complex

## Real-Time Example: Heart Disease Detection System

In a heart disease detection system, instead of using a single classifier like a Decision Tree, multiple models such as
Random Forest and Gradient Boosting can be combined using ensemble methods. This improves prediction accuracy and helps
doctors make better decisions by providing reliable risk predictions.

## Conclusion

Ensemble methods are powerful techniques in machine learning that improve model accuracy by combining multiple models.
They are especially useful in sensitive applications such as medical diagnosis, where prediction reliability is
very important. Using ensemble techniques like Bagging, Boosting, and Stacking helps produce better and more stable results compared to individual models.
```

```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)

# Create individual models
model1 = DecisionTreeClassifier()
model2 = SVC(probability=True)
model3 = LogisticRegression()

# Create Ensemble Model (Voting Classifier)
ensemble_model = VotingClassifier(
    estimators=[
        ('dt', model1),
        ('svm', model2),
        ('lr', model3)
    ],
    voting='soft'
)

# Train model
ensemble_model.fit(X_train, y_train)

# Predict results
prediction = ensemble_model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, prediction)

print("Ensemble Model Accuracy:", accuracy)
```
